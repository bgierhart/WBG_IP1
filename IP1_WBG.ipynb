# 4. Import car-sales.csv  into a DataFrame called car_sales and investigate about the dataset.

```python
car_sales=pd.read_csv("car-sales.csv")
car_sales.head()
```

# 5. Remove all the symbols (\$\,)from the 'Price' column and convert  and update it to  64-bit int numbers using astype(int64) Hint. Look for string functions like replace/strip). 

```python
#remove $
car_sales['Price']=car_sales['Price'].str.replace('$','',regex=False)
#remove ,
car_sales['Price']=car_sales['Price'].str.replace(',','',regex=False)
#change from strings to ints
car_sales['Price']=car_sales['Price'].astype(int)
```

```python
#demo of successful replace
car_sales['Price']
```

```python
#demo of successful dtype change
car_sales['Price'].info()
```

# 6. Add a column called 'Total Sales' to car_sales which cumulatively adds the 'Price' column look up the cumsum() function.

```python
car_sales['Total Sales']=car_sales['Price'].cumsum()
#demo of succesful column addition
car_sales.head()
```

# 7.Add a column called 'Sale Date' which lists a series of successive dates starting from today (your today). Check out date_range function and keep the periods=len(car_sales).

```python
start_date=datetime.today()
car_sales['Sale Date']=pd.date_range(start=start_date, periods=len(car_sales), freq='D')
car_sales['Sale Date']=car_sales['Sale Date'].dt.date
#demo of succesful column addition
car_sales.head()
```

# 8. Use the matplotlib line plot function to plot the 'Price' column versus the 'Total Sales' column (Give proper title and appropriate labels.

```python
plt.plot(car_sales['Price'],car_sales['Total Sales'])
#labelling
plt.title('Price vs Total Sales Line Plot')
plt.xlabel('Price')
plt.ylabel('Total Sales')
```

# 9. Create a matplotlib scatter plot of the 'Odometer (KM)' vs 'Price' column.

```python
plt.scatter(car_sales['Odometer (KM)'],car_sales['Price'])
plt.title('Odometer (KM) vs Price Scatterplot')
plt.xlabel('Odometer (KM)')
plt.ylabel('Price')
```

# 10. Create a matplotlib bar graph of the 'Make' vs 'Odometer (KM)' columns

```python
plt.bar(car_sales['Make'],car_sales['Odometer (KM)'])
plt.title('Make vs Odometer (KM) Bar Plot')
plt.xlabel('Make')
plt.ylabel('Odometer (KM)')
```

# 11. Create a histogram of the 'Make' column

```python
plt.hist(car_sales['Make'])
plt.title('Histogram of Make')
plt.xlabel('Make')
plt.ylabel('Count')
```

# 12. Create a subplot of multiple plots ( step 8,9,10,11).

```python
#creation of figure with axes for subplot, constrained layout needed for spacing
figure, axis=plt.subplots(2,2,constrained_layout=True)

axis[0,0].plot(car_sales['Price'],car_sales['Total Sales'])
axis[0,0].set_title("Price vs Total Sales Line Plot")
axis[0,0].set_xlabel('Price')
axis[0,0].set_ylabel('Total Sales')

axis[0,1].scatter(car_sales['Odometer (KM)'],car_sales['Price'])
axis[0,1].set_title('Odometer (KM) vs Price Scatterplot')
axis[0,1].set_xlabel('Odometer (KM)')
axis[0,1].set_ylabel('Price')

axis[1,0].bar(car_sales['Make'],car_sales['Odometer (KM)'])
axis[1,0].set_title('Make vs Odometer (KM) Bar Plot')
axis[1,0].set_xlabel('Make')
axis[1,0].set_ylabel('Odometer (KM)')

axis[1,1].hist(car_sales['Make'])
axis[1,1].set_title('Histogram of Make')
axis[1,1].set_xlabel('Make')
axis[1,1].set_ylabel('Count')
```

# 16.	Find out Pearson correlation between 'Total Sales','Price','Odometer (KM)'. Comment on the correlation 

```python
#creates matrix for heatmap
corr_matrix=car_sales[['Total Sales', 'Price', 'Odometer (KM)']].corr()
#annot=True adds number of Pearson correlation
sns.heatmap(corr_matrix, annot=True)
```

#### There is a strong negative correlation between Price and Odometer (KM). The lower the price, the more likely the odometer is to be high and vice versa.
